# Find reactions/rows shared between the test sets

```python
import pandas as pd
```

```python
no_trust_test_path = "/Users/dsw46/Library/CloudStorage/OneDrive-UniversityofCambridge/Datasets/orderly_v5/orderly_no_trust_no_map_test.parquet"
with_trust_test_path = "/Users/dsw46/Library/CloudStorage/OneDrive-UniversityofCambridge/Datasets/orderly_v5/orderly_with_trust_no_map_test.parquet"
df1 = pd.read_parquet(no_trust_test_path)
df2 = pd.read_parquet(with_trust_test_path)
```

```python
# Find indices where the rxn_str is the same

# Extract the values from the "rxn_str" column in df1 and df2
rxn_str_2 = df2["rxn_str"].values

# Filter df1 based on values present in df2 and vice versa
intersection_indices1 = df1[df1["rxn_str"].isin(rxn_str_2)].index

# Print the filtered rows
print(len(intersection_indices1))
print(len(df1))
print(len(df2))

corresponding_indices_in_df2 = []

# Loop through the filtered indices in df1
for idx in intersection_indices1:
    # Extract the 'rxn_str' value from df1
    rxn_str_value = df1.loc[idx, 'rxn_str']
    
    # Find the index in df2 where 'rxn_str' equals rxn_str_value
    corresponding_indices_in_df2.append(df2[df2['rxn_str'] == rxn_str_value].index[0])
```
